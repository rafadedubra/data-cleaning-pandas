```python
null_cols = data.isnull().sum()
null_cols
```

```python
data = data.drop(["Unnamed: 22", "Unnamed: 23","Case Number.1","Case Number.2","Time",
                  "href formula",'pdf','original order', "href", "Area", "Location", "Name", "Time", "Species ", "Investigator or Source", "pdf", "Injury", "Date", "Age"], axis=1)
```

```python
data.head(15)
```

```python
len(data)
```

```python
data = data.loc[data['Type'] == 'Unprovoked']
data = data.loc[(data['Activity'] == "Surfing") | (data['Activity'] == "Swimming") | (data['Activity'] == "Fishing") | (data['Activity'] == "Spearfishing") | (data['Activity'] == "Bathing")]
data
```

```python
data.columns = data.columns.str.strip()
data['Sex']
```

```python
set(data['Sex'])
```

```python
data['Sex'] = data['Sex'].str.strip()
data['Sex'] = data['Sex'].str.replace('N', 'M')
data['Sex'] = data['Sex'].str.replace('lli', 'Unknown')
data['Sex'] = data['Sex'].fillna('Unknown')
data['Sex'] = data['Sex'].str.replace('.', 'Unknown')
```

```python
set(data['Sex'])
```

```python
set(data['Country'])
```

```python
set(data['Fatal (Y/N)'])
```

```python
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.strip()
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('n', 'N')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('y', 'Y')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('F', 'Y')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].str.replace('#VALUE!', 'U')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].fillna('U')

data = data.loc[data['Fatal (Y/N)'] != 'UNKNOWN']
```

```python
set(data['Year'])
data["Year"] = data["Year"].fillna(0.0).astype(int)
data = data.loc[data['Year'] != 0]

set(data['Year'])
```

```python
data = data.dropna(how='all',axis=0) 
data.head()
```

```python
data.to_csv('./data/attacks_clean.csv', index = False)
```
